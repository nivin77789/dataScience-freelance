```python
# Fill missing values in 'new_capacity' based on mean of each 'capacity_units' group
df_prod1['new_capacity'] = df_prod1.groupby('capacity_units')['new_capacity'].transform(
    lambda x: x.fillna(x.mean())
)
```

```python
len(df_prod1['capacity_units'].unique())
```

```python
# Step 1: Check how many unique capacity units exist
print("Unique capacity units:", df_prod1['capacity_units'].nunique())

# Step 2: Compute mean 'new_capacity' for each 'capacity_units'
capacity_means = df_prod1.groupby('capacity_units')['new_capacity'].mean()

# Step 3: Display the computed means (optional)
print("\nMean new_capacity for each capacity unit:")
print(capacity_means)

# Step 4: Impute missing 'new_capacity' based on corresponding capacity unit mean
df_prod1['new_capacity'] = df_prod1.apply(
    lambda row: capacity_means[row['capacity_units']]
    if pd.isna(row['new_capacity']) and row['capacity_units'] in capacity_means
    else row['new_capacity'],
    axis=1
)

# Step 5: Verify missing values after imputation
print("\nMissing values in 'new_capacity' after imputation:",
      df_prod1['new_capacity'].isna().sum())
```

```python
# Count of rows where both new_capacity and capacity_units are missing
both_null_count = df_prod1[df_prod1['new_capacity'].isna() & df_prod1['capacity_units'].isna()].shape[0]

print("Number of rows where both 'new_capacity' and 'capacity_units' are null:", both_null_count)
```

```python
# Drop rows where 'new_capacity' is null
df_prod1 = df_prod1.dropna(subset=['new_capacity'])

# Verify the result
print("Remaining rows after dropping null 'new_capacity':", df_prod1.shape[0])
print("Missing values in 'new_capacity' after dropping:", df_prod1['new_capacity'].isna().sum())
```

```python
import pandas as pd

# Copy dataframe
df = df_prod1.copy()

# 1 Basic Info
print("Dataset Shape:", df.shape)
print("\nData Types:\n", df.dtypes)

# 2 Missing Values
missing_count = df.isnull().sum()
missing_percent = (df.isnull().mean() * 100).round(2)
missing_summary = pd.DataFrame({
    'Column': df.columns,
    'Missing Values': missing_count,
    '% Missing': missing_percent
})
print("\nMissing Values Summary:\n", missing_summary)

# 3 Duplicates
total_duplicates = df.duplicated().sum()
print("\nTotal duplicate rows:", total_duplicates)

# Duplicates based on key identifiers
key_duplicates = df.duplicated(subset=['company', 'project_name', 'project_number']).sum()
print("Duplicates by company + project_name + project_number:", key_duplicates)

# 4 Object Columns Analysis
print("\nObject Columns Analysis:")
object_cols = df.select_dtypes(include='object').columns

for col in object_cols:
    print(f"\nColumn: {col}")
    
    # Unique values and counts
    value_counts = df[col].value_counts(dropna=False)
    print("Top 10 Value Counts:\n", value_counts.head(10))
    
    # Strip spaces and lowercase to detect inconsistencies
    clean_col = col + '_clean'
    df[clean_col] = df[col].astype(str).str.strip().str.lower()
    unique_cleaned = df[clean_col].nunique()
    print("Unique values after cleaning (strip/lowercase):", unique_cleaned)
    
    # Detect non-alphanumeric characters for text columns
    # Skip numeric-like columns like product_code
    if not col.endswith('code') and col not in ['capacity_units']:
        mask_nonalpha = df[clean_col].str.contains(r'[^a-z0-9\s&,-]', regex=True, na=False)
        if mask_nonalpha.sum() > 0:
            print("Non-alphanumeric entries:", df.loc[mask_nonalpha, col].unique())

# 5 Numeric Columns Analysis
print("\nNumeric Columns Analysis:")
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in numeric_cols:
    print(f"\nColumn: {col}")
    print("Missing values:", df[col].isna().sum())
    print("Unique values:", df[col].nunique())
    print("Statistical summary:\n", df[col].describe())

# 6 Rows where both 'new_capacity' and 'capacity_units' are null
both_null_count = df[df['new_capacity'].isna() & df['capacity_units'].isna()].shape[0]
print("\nRows where both 'new_capacity' and 'capacity_units' are missing:", both_null_count)

# 7 Flag expansion projects (use correct column names)
df['is_expansion_project'] = df['expansion_from'].apply(lambda x: 0 if pd.isna(x) else 1)
print("\nFlagged expansion projects. Sample:")
print(df[['project_name', 'expansion_from', 'expansion_to', 'is_expansion_project']].head())
```
